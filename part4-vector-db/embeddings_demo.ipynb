```python
import numpy as np

# 1. Load at least 5 product descriptions (from our dataset)
products = ["Laptop", "Headphones", "Notebook", "Pen Set", "Desk Chair"]

# 2. Generate mock embeddings (lists of 4 random floats)
# In real life, these would come from an AI model like OpenAI or BERT
embeddings = {
    "Laptop": [0.91, 0.12, 0.05, 0.08],
    "Headphones": [0.88, 0.15, 0.09, 0.12],
    "Notebook": [0.02, 0.85, 0.11, 0.05],
    "Pen Set": [0.05, 0.82, 0.08, 0.02],
    "Desk Chair": [0.12, 0.09, 0.77, 0.88]
}

# 3. Calculate Cosine Similarity between "Laptop" and "Headphones"
# Formula: (A dot B) / (||A|| * ||B||)
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

vec_a = embeddings["Laptop"]
vec_b = embeddings["Headphones"]
similarity = cosine_similarity(vec_a, vec_b)

print(f"Similarity between Laptop and Headphones: {similarity:.4f}")

# 4. Usefulness for Recommendation Engines:
# This is useful because it allows a system to suggest products based on
# "meaning" rather than just categories. If a user buys a Laptop, the system
# can find items with high similarity (like Headphones) to recommend next,
# improving the shopping experience through semantic relationships.
```
